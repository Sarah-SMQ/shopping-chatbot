```python
# ---------------- محاكاة البيانات ----------------
query = "iphone 15 or iphone 16"

products_by_item = {
    "iphone 15 or iphone 16": [
        {"title": "iPhone 15"},
        {"title": "iPhone 16"},
        {"title": "iPhone 16 Plus"},
        {"title": "iPhone 16 Pro"},
        {"title": "iPhone 16 5G"},
    ]
}

ai_reply = """
المقارنة بين iPhone 15 و iPhone 16:
iPhone 15 ممتاز للميزانية، iPhone 16 أحدث مع A18 وميزات محسّنة.
"""

# ---------------- دالة حساب الدقة ----------------
def evaluate_accuracy_three(Q, context, final_answer):
    if not context:
        return 0

    relevant = sum(1 for c in context if any(word.lower() in c["title"].lower() for word in Q.split()))
    relevance_score = relevant / len(context)

    faithful = sum(1 for c in context if c["title"].lower() in final_answer.lower())
    faithfulness_score = faithful / len(context)

    mentioned = len(set([c["title"] for c in context if c["title"].lower() in final_answer.lower()]))
    completeness_score = min(1, mentioned / len(context))

    total_score = round((0.5 * faithfulness_score + 0.3 * relevance_score + 0.2 * completeness_score) * 100, 2)
    return total_score

flat_context = [p for plist in products_by_item.values() for p in plist]

score = evaluate_accuracy_three(query, flat_context, ai_reply)
print(f"🎯 True Evaluation Score: {score}%")
```

```python
import os
import requests
import json
from dotenv import load_dotenv

# ---------------- Load environment ----------------
load_dotenv(r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\.env")

SERPAPI_KEY = os.getenv("SERPAPI_KEY")
GROQ_KEY = os.getenv("GROQ_KEY")
GROQ_URL = os.getenv("GROQ_URL")
GROQ_MODEL = os.getenv("GROQ_MODEL")

# ---------------- Test SerpAPI ----------------
def test_serpapi(query="Apple Watch sapphire"):
    if not SERPAPI_KEY:
        print("❌ SERPAPI_KEY missing")
        return
    try:
        url = "https://serpapi.com/search.json"
        params = {
            "engine": "google_shopping",
            "q": query,
            "hl": "en",
            "gl": "us",
            "api_key": SERPAPI_KEY,
            "tbm": "shop",
        }
        response = requests.get(url, params=params, timeout=10)
        response.raise_for_status()
        data = response.json()
        results = data.get("shopping_results", [])
        print(f"✅ SerpAPI connected successfully! Found {len(results)} results.")
    except Exception as e:
        print(f"❌ SerpAPI error: {e}")

# ---------------- Test Groq API ----------------
def test_groq():
    if not GROQ_KEY or not GROQ_URL or not GROQ_MODEL:
        print("❌ GROQ environment variables missing")
        return
    try:
        headers = {
            "Authorization": f"Bearer {GROQ_KEY}",
            "Content-Type": "application/json",
        }
        payload = {
            "model": GROQ_MODEL,
            "messages": [
                {"role": "system", "content": "You are a test assistant."},
                {"role": "user", "content": "Hello, just testing the connection."},
            ],
        }
        response = requests.post(GROQ_URL, headers=headers, json=payload, timeout=10)
        response.raise_for_status()
        data = response.json()
        if "choices" in data:
            print("✅ Groq API connected successfully!")
        else:
            print("❌ Groq response missing 'choices':", data)
    except Exception as e:
        print(f"❌ Groq API error: {e}")

# ---------------- Run tests ----------------
print("Testing SerpAPI...")
test_serpapi()

print("\nTesting Groq API...")
test_groq()
```

```python
import json
import re
from googletrans import Translator

JSON_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping.json"

with open(JSON_FILE, "r", encoding="utf-8") as f:
    data = json.load(f)

translator = Translator()

def translate_markdown_table(text):
    """
    يترجم نص يحتوي على جدول Markdown مع الحفاظ على الخطوط | و - 
    """
    lines = text.split("\n")
    new_lines = []
    for line in lines:
        if re.match(r"^\s*\|", line):  # سطر جدول
            # تقسيم الخلايا، تجاهل | في البداية والنهاية
            cells = [c.strip() for c in line.strip("|").split("|")]
            translated_cells = [translator.translate(c, src="ar", dest="en").text if c else "" for c in cells]
            new_line = "| " + " | ".join(translated_cells) + " |"
            new_lines.append(new_line)
        else:
            # ترجمة السطر العادي
            new_lines.append(translator.translate(line, src="ar", dest="en").text if line.strip() else line)
    return "\n".join(new_lines)

for entry in data:
    arabic_text = entry.get("ai_reply", "")
    if arabic_text:
        entry["ai_reply"] = translate_markdown_table(arabic_text)

# حفظ على نفس الملف
with open(JSON_FILE, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("✅ All ai_reply fields translated to English while preserving Markdown tables.")
```

```python
import json
import pandas as pd

# مسار ملف JSON
JSON_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping.json"
# مسار ملف Excel النهائي
EXCEL_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval.xlsx"

# قراءة JSON
with open(JSON_FILE, "r", encoding="utf-8") as f:
    data = json.load(f)

rows = []
for item in data:
    query = item.get("query", "")
    ai_reply = item.get("ai_reply", "")
    
    # تلخيص المحتوى في عمود واحد كنص
    products = item.get("products", [])
    content_list = []
    for p in products:
        title = p.get("title", "")
        price = p.get("price", "")
        source = p.get("source", "")
        content_list.append(f"{title} | {price} | {source}")
    content_text = "\n".join(content_list)
    
    # التقييم
    eval_score = item.get("evaluation_score", {})
    faithfulness = eval_score.get("faithfulness", "")
    relevance = eval_score.get("relevance", "")
    completeness = eval_score.get("completeness", "")
    total = eval_score.get("total", "")

    # إضافة صف
    rows.append({
        "query": query,
        "content": content_text,
        "ai_reply": ai_reply,
        "faithfulness": faithfulness,
        "relevance": relevance,
        "completeness": completeness,
        "total": total
    })

# تحويل للقالب DataFrame وحفظ Excel
df = pd.DataFrame(rows)
df.to_excel(EXCEL_FILE, index=False)

print(f"تم تحويل JSON إلى Excel بنجاح: {EXCEL_FILE}")
```

```python
import pandas as pd
from difflib import SequenceMatcher

# ------------------- المسارات -------------------
EXCEL_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval.xlsx"
REPORT_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval_report.xlsx"

# ------------------- قراءة الملف -------------------
df = pd.read_excel(EXCEL_FILE)

# ------------------- نقطة 4: تحليل العلاقة بين المعايير -------------------
correlation = df[['faithfulness', 'relevance', 'completeness', 'total']].corr()
print("Correlation Matrix:")
print(correlation)

# ------------------- نقطة 5: مقارنة ai_reply بالمحتوى -------------------
def similarity_score(a, b):
    # التأكد من أن القيم ليست فارغة أو NaN
    if pd.isna(a) or pd.isna(b):
        return 0
    # تحويل لأي نوع نصي
    a = str(a)
    b = str(b)
    return round(SequenceMatcher(None, a, b).ratio() * 100, 2)

df['similarity_score'] = df.apply(lambda row: similarity_score(row['ai_reply'], row['content']), axis=1)

# ------------------- تصنيف الأسئلة حسب الدرجة الكلية -------------------
def classify_total(total):
    if pd.isna(total):
        return 'Unknown'
    elif total >= 80:
        return 'High'
    elif total >= 50:
        return 'Medium'
    else:
        return 'Low'

df['performance_class'] = df['total'].apply(classify_total)

# ------------------- تقرير ملخص -------------------
summary = {
    'total_questions': len(df),
    'high_quality': len(df[df['performance_class'] == 'High']),
    'medium_quality': len(df[df['performance_class'] == 'Medium']),
    'low_quality': len(df[df['performance_class'] == 'Low']),
    'best_5_queries': df.sort_values(by='total', ascending=False).head(5)['query'].tolist(),
    'worst_5_queries': df.sort_values(by='total').head(5)['query'].tolist(),
    'average_scores': {
        'faithfulness': df['faithfulness'].mean(),
        'relevance': df['relevance'].mean(),
        'completeness': df['completeness'].mean(),
        'total': df['total'].mean(),
        'similarity_score': df['similarity_score'].mean()
    },
    'correlation_matrix': correlation.to_dict()
}

print("\nSummary:")
print(summary)

# ------------------- حفظ التقرير -------------------
with pd.ExcelWriter(REPORT_FILE) as writer:
    df.to_excel(writer, index=False, sheet_name='Evaluated Data')
    pd.DataFrame(correlation).to_excel(writer, sheet_name='Correlation Matrix')
    # حفظ تقرير الملخص في sheet منفصل
    pd.DataFrame.from_dict(summary, orient='index').to_excel(writer, sheet_name='Summary')

print(f"\nتم إنشاء التقرير مع التحليلات: {REPORT_FILE}")
```

```python
import os
import json
import pandas as pd
from difflib import SequenceMatcher
import requests
from dotenv import load_dotenv
import time

# ------------------ إعداد المتغيرات ------------------
load_dotenv(r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\.env")
GROQ_KEY = os.getenv("GROQ_KEY")
GROQ_URL = os.getenv("GROQ_URL")
GROQ_MODEL = os.getenv("GROQ_MODEL")

EXCEL_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval.xlsx"
CHECKPOINT_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval_checkpoint.xlsx"
NEW_REPORT_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval_comparison.xlsx"

# ------------------ دوال Groq ------------------
def call_groq(messages):
    headers = {
        "Authorization": f"Bearer {GROQ_KEY}",
        "Content-Type": "application/json",
    }
    payload = {"model": GROQ_MODEL, "messages": messages}
    response = requests.post(GROQ_URL, headers=headers, json=payload)
    if response.status_code != 200:
        raise RuntimeError(f"GROQ API error {response.status_code}: {response.text}")
    data = response.json()
    return data["choices"][0]["message"]["content"]

def evaluate_row_with_groq(query, content, ai_reply, max_retries=5):
    prompt = f"""
You are an expert AI evaluator.
Question: {query}
Available content:
{content}
AI's reply:
{ai_reply}

Rate Faithfulness, Relevance, Completeness from 10 to 100 each.
Provide JSON: {{"faithfulness":..., "relevance":..., "completeness":..., "total":...}}
"""
    for attempt in range(max_retries):
        try:
            resp = call_groq([
                {"role": "system", "content": "You are an AI answer evaluation assistant."},
                {"role": "user", "content": prompt}
            ])
            scores = json.loads(resp)
            for k in ["faithfulness", "relevance", "completeness"]:
                val = scores.get(k, 10)
                scores[k] = max(10, min(val, 100))
            total = round(0.4*scores["faithfulness"] + 0.3*scores["relevance"] + 0.3*scores["completeness"], 2)
            scores["total"] = max(10, min(total, 100))
            scores["error"] = False
            return scores
        except Exception as e:
            wait_time = 5 + attempt * 2  # 5 ثواني + زيادة تدريجية
            print(f"Attempt {attempt+1} failed: {e}. Retrying in {wait_time}s...")
            time.sleep(wait_time)
    print(f"Evaluation failed for query: {query}. Marking as error.")
    return {"faithfulness":10, "relevance":10, "completeness":10, "total":10, "error": True}

# ------------------ دوال مساعدة ------------------
def similarity_score(a, b):
    if not a or not b:
        return 0
    return round(SequenceMatcher(None, a, b).ratio()*100,2)

def classify_total(total):
    if total >= 80:
        return 'High'
    elif total >= 50:
        return 'Medium'
    else:
        return 'Low'

# ------------------ قراءة الملف ------------------
df = pd.read_excel(EXCEL_FILE)

# التأكد من الأعمدة الجديدة
for col in ['new_faithfulness','new_relevance','new_completeness','new_total','error']:
    if col not in df.columns:
        df[col] = None

# إعادة تحميل checkpoint إذا وجد
if os.path.exists(CHECKPOINT_FILE):
    df_checkpoint = pd.read_excel(CHECKPOINT_FILE)
    df.update(df_checkpoint)

# ------------------ تقييم الصفوف ------------------
total_rows = len(df)
for idx, row in df.iterrows():
    if pd.notna(row['new_total']) and not row.get('error', False):
        continue  # تخطي الصفوف التي تم تقييمها بدون خطأ
    scores = evaluate_row_with_groq(row['query'], row['content'], row['ai_reply'])
    df.at[idx,'new_faithfulness'] = scores['faithfulness']
    df.at[idx,'new_relevance'] = scores['relevance']
    df.at[idx,'new_completeness'] = scores['completeness']
    df.at[idx,'new_total'] = scores['total']
    df.at[idx,'error'] = scores.get('error', False)
    
    # حفظ checkpoint بعد كل صف
    df.to_excel(CHECKPOINT_FILE, index=False)
    print(f"Processed row {idx+1}/{total_rows}")
    time.sleep(0.5)

# ------------------ الفروقات والتقييم النهائي ------------------
df['diff_faithfulness'] = df['new_faithfulness'] - df['faithfulness']
df['diff_relevance'] = df['new_relevance'] - df['relevance']
df['diff_completeness'] = df['new_completeness'] - df['completeness']
df['diff_total'] = df['new_total'] - df['total']
df['similarity_score'] = df.apply(lambda row: similarity_score(row['ai_reply'], row['content']), axis=1)
df['new_performance_class'] = df['new_total'].apply(classify_total)

# ------------------ تحليل العلاقة ------------------
correlation = df[['new_faithfulness','new_relevance','new_completeness','new_total']].corr()

# ------------------ حفظ التقرير النهائي ------------------
df.to_excel(NEW_REPORT_FILE, index=False)
pd.DataFrame(correlation).to_excel(NEW_REPORT_FILE.replace('.xlsx','_correlation.xlsx'))
print(f"\nتم إنشاء التقرير النهائي: {NEW_REPORT_FILE}")
```
