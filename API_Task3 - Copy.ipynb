```python
import json
import re
from googletrans import Translator

JSON_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping.json"

with open(JSON_FILE, "r", encoding="utf-8") as f:
    data = json.load(f)

translator = Translator()

def translate_markdown_table(text):
    """
    Ÿäÿ™ÿ±ÿ¨ŸÖ ŸÜÿµ Ÿäÿ≠ÿ™ŸàŸä ÿπŸÑŸâ ÿ¨ÿØŸàŸÑ Markdown ŸÖÿπ ÿßŸÑÿ≠ŸÅÿßÿ∏ ÿπŸÑŸâ ÿßŸÑÿÆÿ∑Ÿàÿ∑ | Ÿà - 
    """
    lines = text.split("\n")
    new_lines = []
    for line in lines:
        if re.match(r"^\s*\|", line):  # ÿ≥ÿ∑ÿ± ÿ¨ÿØŸàŸÑ
            # ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿÆŸÑÿßŸäÿßÿå ÿ™ÿ¨ÿßŸáŸÑ | ŸÅŸä ÿßŸÑÿ®ÿØÿßŸäÿ© ŸàÿßŸÑŸÜŸáÿßŸäÿ©
            cells = [c.strip() for c in line.strip("|").split("|")]
            translated_cells = [translator.translate(c, src="ar", dest="en").text if c else "" for c in cells]
            new_line = "| " + " | ".join(translated_cells) + " |"
            new_lines.append(new_line)
        else:
            # ÿ™ÿ±ÿ¨ŸÖÿ© ÿßŸÑÿ≥ÿ∑ÿ± ÿßŸÑÿπÿßÿØŸä
            new_lines.append(translator.translate(line, src="ar", dest="en").text if line.strip() else line)
    return "\n".join(new_lines)

for entry in data:
    arabic_text = entry.get("ai_reply", "")
    if arabic_text:
        entry["ai_reply"] = translate_markdown_table(arabic_text)

# ÿ≠ŸÅÿ∏ ÿπŸÑŸâ ŸÜŸÅÿ≥ ÿßŸÑŸÖŸÑŸÅ
with open(JSON_FILE, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("‚úÖ All ai_reply fields translated to English while preserving Markdown tables.")
```

```python
import pandas as pd
from difflib import SequenceMatcher

# ------------------- ÿßŸÑŸÖÿ≥ÿßÿ±ÿßÿ™ -------------------
EXCEL_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval.xlsx"
REPORT_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval_report.xlsx"

# ------------------- ŸÇÿ±ÿßÿ°ÿ© ÿßŸÑŸÖŸÑŸÅ -------------------
df = pd.read_excel(EXCEL_FILE)

# ------------------- ŸÜŸÇÿ∑ÿ© 4: ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿπŸÑÿßŸÇÿ© ÿ®ŸäŸÜ ÿßŸÑŸÖÿπÿßŸäŸäÿ± -------------------
correlation = df[['faithfulness', 'relevance', 'completeness', 'total']].corr()
print("Correlation Matrix:")
print(correlation)

# ------------------- ŸÜŸÇÿ∑ÿ© 5: ŸÖŸÇÿßÿ±ŸÜÿ© ai_reply ÿ®ÿßŸÑŸÖÿ≠ÿ™ŸàŸâ -------------------
def similarity_score(a, b):
    # ÿßŸÑÿ™ÿ£ŸÉÿØ ŸÖŸÜ ÿ£ŸÜ ÿßŸÑŸÇŸäŸÖ ŸÑŸäÿ≥ÿ™ ŸÅÿßÿ±ÿ∫ÿ© ÿ£Ÿà NaN
    if pd.isna(a) or pd.isna(b):
        return 0
    # ÿ™ÿ≠ŸàŸäŸÑ ŸÑÿ£Ÿä ŸÜŸàÿπ ŸÜÿµŸä
    a = str(a)
    b = str(b)
    return round(SequenceMatcher(None, a, b).ratio() * 100, 2)

df['similarity_score'] = df.apply(lambda row: similarity_score(row['ai_reply'], row['content']), axis=1)

# ------------------- ÿ™ÿµŸÜŸäŸÅ ÿßŸÑÿ£ÿ≥ÿ¶ŸÑÿ© ÿ≠ÿ≥ÿ® ÿßŸÑÿØÿ±ÿ¨ÿ© ÿßŸÑŸÉŸÑŸäÿ© -------------------
def classify_total(total):
    if pd.isna(total):
        return 'Unknown'
    elif total >= 80:
        return 'High'
    elif total >= 50:
        return 'Medium'
    else:
        return 'Low'

df['performance_class'] = df['total'].apply(classify_total)

# ------------------- ÿ™ŸÇÿ±Ÿäÿ± ŸÖŸÑÿÆÿµ -------------------
summary = {
    'total_questions': len(df),
    'high_quality': len(df[df['performance_class'] == 'High']),
    'medium_quality': len(df[df['performance_class'] == 'Medium']),
    'low_quality': len(df[df['performance_class'] == 'Low']),
    'best_5_queries': df.sort_values(by='total', ascending=False).head(5)['query'].tolist(),
    'worst_5_queries': df.sort_values(by='total').head(5)['query'].tolist(),
    'average_scores': {
        'faithfulness': df['faithfulness'].mean(),
        'relevance': df['relevance'].mean(),
        'completeness': df['completeness'].mean(),
        'total': df['total'].mean(),
        'similarity_score': df['similarity_score'].mean()
    },
    'correlation_matrix': correlation.to_dict()
}

print("\nSummary:")
print(summary)

# ------------------- ÿ≠ŸÅÿ∏ ÿßŸÑÿ™ŸÇÿ±Ÿäÿ± -------------------
with pd.ExcelWriter(REPORT_FILE) as writer:
    df.to_excel(writer, index=False, sheet_name='Evaluated Data')
    pd.DataFrame(correlation).to_excel(writer, sheet_name='Correlation Matrix')
    # ÿ≠ŸÅÿ∏ ÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑŸÖŸÑÿÆÿµ ŸÅŸä sheet ŸÖŸÜŸÅÿµŸÑ
    pd.DataFrame.from_dict(summary, orient='index').to_excel(writer, sheet_name='Summary')

print(f"\nÿ™ŸÖ ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑÿ™ŸÇÿ±Ÿäÿ± ŸÖÿπ ÿßŸÑÿ™ÿ≠ŸÑŸäŸÑÿßÿ™: {REPORT_FILE}")
```

```python
import os
import json
import pandas as pd
from difflib import SequenceMatcher
import requests
from dotenv import load_dotenv
import time

# ------------------ ÿ•ÿπÿØÿßÿØ ÿßŸÑŸÖÿ™ÿ∫Ÿäÿ±ÿßÿ™ ------------------
load_dotenv(r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\.env")
GROQ_KEY = os.getenv("GROQ_KEY")
GROQ_URL = os.getenv("GROQ_URL")
GROQ_MODEL = os.getenv("GROQ_MODEL")

EXCEL_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval.xlsx"
CHECKPOINT_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval_checkpoint.xlsx"
NEW_REPORT_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval_comparison.xlsx"

# ------------------ ÿØŸàÿßŸÑ Groq ------------------
def call_groq(messages):
    headers = {
        "Authorization": f"Bearer {GROQ_KEY}",
        "Content-Type": "application/json",
    }
    payload = {"model": GROQ_MODEL, "messages": messages}
    response = requests.post(GROQ_URL, headers=headers, json=payload)
    if response.status_code != 200:
        raise RuntimeError(f"GROQ API error {response.status_code}: {response.text}")
    data = response.json()
    return data["choices"][0]["message"]["content"]

def evaluate_row_with_groq(query, content, ai_reply, max_retries=5):
    prompt = f"""
You are an expert AI evaluator.
Question: {query}
Available content:
{content}
AI's reply:
{ai_reply}

Rate Faithfulness, Relevance, Completeness from 10 to 100 each.
Provide JSON: {{"faithfulness":..., "relevance":..., "completeness":..., "total":...}}
"""
    for attempt in range(max_retries):
        try:
            resp = call_groq([
                {"role": "system", "content": "You are an AI answer evaluation assistant."},
                {"role": "user", "content": prompt}
            ])
            scores = json.loads(resp)
            for k in ["faithfulness", "relevance", "completeness"]:
                val = scores.get(k, 10)
                scores[k] = max(10, min(val, 100))
            total = round(0.4*scores["faithfulness"] + 0.3*scores["relevance"] + 0.3*scores["completeness"], 2)
            scores["total"] = max(10, min(total, 100))
            scores["error"] = False
            return scores
        except Exception as e:
            wait_time = 5 + attempt * 2  # 5 ÿ´ŸàÿßŸÜŸä + ÿ≤ŸäÿßÿØÿ© ÿ™ÿØÿ±Ÿäÿ¨Ÿäÿ©
            print(f"Attempt {attempt+1} failed: {e}. Retrying in {wait_time}s...")
            time.sleep(wait_time)
    print(f"Evaluation failed for query: {query}. Marking as error.")
    return {"faithfulness":10, "relevance":10, "completeness":10, "total":10, "error": True}

# ------------------ ÿØŸàÿßŸÑ ŸÖÿ≥ÿßÿπÿØÿ© ------------------
def similarity_score(a, b):
    if not a or not b:
        return 0
    return round(SequenceMatcher(None, a, b).ratio()*100,2)

def classify_total(total):
    if total >= 80:
        return 'High'
    elif total >= 50:
        return 'Medium'
    else:
        return 'Low'

# ------------------ ŸÇÿ±ÿßÿ°ÿ© ÿßŸÑŸÖŸÑŸÅ ------------------
df = pd.read_excel(EXCEL_FILE)

# ÿßŸÑÿ™ÿ£ŸÉÿØ ŸÖŸÜ ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿßŸÑÿ¨ÿØŸäÿØÿ©
for col in ['new_faithfulness','new_relevance','new_completeness','new_total','error']:
    if col not in df.columns:
        df[col] = None

# ÿ•ÿπÿßÿØÿ© ÿ™ÿ≠ŸÖŸäŸÑ checkpoint ÿ•ÿ∞ÿß Ÿàÿ¨ÿØ
if os.path.exists(CHECKPOINT_FILE):
    df_checkpoint = pd.read_excel(CHECKPOINT_FILE)
    df.update(df_checkpoint)

# ------------------ ÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿµŸÅŸàŸÅ ------------------
total_rows = len(df)
for idx, row in df.iterrows():
    if pd.notna(row['new_total']) and not row.get('error', False):
        continue  # ÿ™ÿÆÿ∑Ÿä ÿßŸÑÿµŸÅŸàŸÅ ÿßŸÑÿ™Ÿä ÿ™ŸÖ ÿ™ŸÇŸäŸäŸÖŸáÿß ÿ®ÿØŸàŸÜ ÿÆÿ∑ÿ£
    scores = evaluate_row_with_groq(row['query'], row['content'], row['ai_reply'])
    df.at[idx,'new_faithfulness'] = scores['faithfulness']
    df.at[idx,'new_relevance'] = scores['relevance']
    df.at[idx,'new_completeness'] = scores['completeness']
    df.at[idx,'new_total'] = scores['total']
    df.at[idx,'error'] = scores.get('error', False)
    
    # ÿ≠ŸÅÿ∏ checkpoint ÿ®ÿπÿØ ŸÉŸÑ ÿµŸÅ
    df.to_excel(CHECKPOINT_FILE, index=False)
    print(f"Processed row {idx+1}/{total_rows}")
    time.sleep(0.5)

# ------------------ ÿßŸÑŸÅÿ±ŸàŸÇÿßÿ™ ŸàÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑŸÜŸáÿßÿ¶Ÿä ------------------
df['diff_faithfulness'] = df['new_faithfulness'] - df['faithfulness']
df['diff_relevance'] = df['new_relevance'] - df['relevance']
df['diff_completeness'] = df['new_completeness'] - df['completeness']
df['diff_total'] = df['new_total'] - df['total']
df['similarity_score'] = df.apply(lambda row: similarity_score(row['ai_reply'], row['content']), axis=1)
df['new_performance_class'] = df['new_total'].apply(classify_total)

# ------------------ ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿπŸÑÿßŸÇÿ© ------------------
correlation = df[['new_faithfulness','new_relevance','new_completeness','new_total']].corr()

# ------------------ ÿ≠ŸÅÿ∏ ÿßŸÑÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑŸÜŸáÿßÿ¶Ÿä ------------------
df.to_excel(NEW_REPORT_FILE, index=False)
pd.DataFrame(correlation).to_excel(NEW_REPORT_FILE.replace('.xlsx','_correlation.xlsx'))
print(f"\nÿ™ŸÖ ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑŸÜŸáÿßÿ¶Ÿä: {NEW_REPORT_FILE}")
```

```python
import os
import json
import pandas as pd
from difflib import SequenceMatcher
import requests
from dotenv import load_dotenv
import time

# ------------------ ÿ•ÿπÿØÿßÿØ ÿßŸÑŸÖÿ™ÿ∫Ÿäÿ±ÿßÿ™ ------------------
load_dotenv(r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\.env")
GROQ_KEY = os.getenv("GROQ_KEY")
GROQ_URL = os.getenv("GROQ_URL")
GROQ_MODEL = os.getenv("GROQ_MODEL")

EXCEL_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval.xlsx"
NEW_REPORT_FILE = r"C:\Users\SarahAlqahtani\Documents\SerpAPI_Research\data_shopping_eval_comparison.xlsx"

# ------------------ ÿØŸàÿßŸÑ Groq ------------------
def call_groq(messages):
    headers = {
        "Authorization": f"Bearer {GROQ_KEY}",
        "Content-Type": "application/json",
    }
    payload = {"model": GROQ_MODEL, "messages": messages}
    response = requests.post(GROQ_URL, headers=headers, json=payload)
    if response.status_code != 200:
        raise RuntimeError(f"GROQ API error {response.status_code}: {response.text}")
    data = response.json()
    return data["choices"][0]["message"]["content"]

def evaluate_row_with_groq(query, content, ai_reply, max_retries=5):
    prompt = f"""
You are an expert AI evaluator.
Question: {query}
Available content:
{content}
AI's reply:
{ai_reply}

Rate Faithfulness, Relevance, Completeness from 10 to 100 each.
Provide JSON: {{"faithfulness":..., "relevance":..., "completeness":..., "total":...}}
"""
    for attempt in range(max_retries):
        try:
            resp = call_groq([
                {"role": "system", "content": "You are an AI answer evaluation assistant."},
                {"role": "user", "content": prompt}
            ])
            scores = json.loads(resp)
            for k in ["faithfulness", "relevance", "completeness"]:
                val = scores.get(k, 10)
                scores[k] = max(10, min(val, 100))
            total = round(0.4*scores["faithfulness"] + 0.3*scores["relevance"] + 0.3*scores["completeness"], 2)
            scores["total"] = max(10, min(total, 100))
            scores["error"] = False
            return scores
        except Exception as e:
            wait_time = 5 + attempt * 2
            print(f"Attempt {attempt+1} failed: {e}. Retrying in {wait_time}s...")
            time.sleep(wait_time)
    print(f"Evaluation failed for query: {query}. Marking as error.")
    return {"faithfulness":10, "relevance":10, "completeness":10, "total":10, "error": True}

# ------------------ ÿØŸàÿßŸÑ ŸÖÿ≥ÿßÿπÿØÿ© ------------------
def similarity_score(a, b):
    if not isinstance(a, str) or not isinstance(b, str):
        return 0
    return round(SequenceMatcher(None, a, b).ratio()*100,2)

def classify_total(total):
    if total >= 80:
        return 'High'
    elif total >= 50:
        return 'Medium'
    else:
        return 'Low'

# ------------------ ŸÇÿ±ÿßÿ°ÿ© ÿßŸÑŸÖŸÑŸÅ ------------------
df = pd.read_excel(EXCEL_FILE)

# ÿßŸÑÿ™ÿ£ŸÉÿØ ŸÖŸÜ ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿßŸÑÿ¨ÿØŸäÿØÿ©
for col in ['new_faithfulness','new_relevance','new_completeness','new_total','error']:
    if col not in df.columns:
        df[col] = None

# ------------------ ÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿµŸÅŸàŸÅ ------------------
total_rows = len(df)
for idx, row in df.iterrows():
    if pd.notna(row['new_total']) and not row.get('error', False):
        continue  # ÿ™ÿÆÿ∑Ÿä ÿßŸÑÿµŸÅŸàŸÅ ÿßŸÑÿ™Ÿä ÿ™ŸÖ ÿ™ŸÇŸäŸäŸÖŸáÿß ÿ®ÿØŸàŸÜ ÿÆÿ∑ÿ£
    scores = evaluate_row_with_groq(row['query'], row['content'], row['ai_reply'])
    df.at[idx,'new_faithfulness'] = scores['faithfulness']
    df.at[idx,'new_relevance'] = scores['relevance']
    df.at[idx,'new_completeness'] = scores['completeness']
    df.at[idx,'new_total'] = scores['total']
    df.at[idx,'error'] = scores.get('error', False)
    
    print(f"Processed row {idx+1}/{total_rows}")
    time.sleep(0.5)

# ------------------ ÿßŸÑŸÅÿ±ŸàŸÇÿßÿ™ ŸàÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑŸÜŸáÿßÿ¶Ÿä ------------------
df['diff_faithfulness'] = df['new_faithfulness'] - df['faithfulness']
df['diff_relevance'] = df['new_relevance'] - df['relevance']
df['diff_completeness'] = df['new_completeness'] - df['completeness']
df['diff_total'] = df['new_total'] - df['total']
df['similarity_score'] = df.apply(lambda row: similarity_score(row['ai_reply'], row['content']), axis=1)
df['new_performance_class'] = df['new_total'].apply(classify_total)

# ------------------ ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿπŸÑÿßŸÇÿ© ------------------
correlation = df[['new_faithfulness','new_relevance','new_completeness','new_total']].corr()

# ------------------ ÿ≠ŸÅÿ∏ ÿßŸÑÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑŸÜŸáÿßÿ¶Ÿä ------------------
df.to_excel(NEW_REPORT_FILE, index=False)
pd.DataFrame(correlation).to_excel(NEW_REPORT_FILE.replace('.xlsx','_correlation.xlsx'))
print(f"\nÿ™ŸÖ ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑŸÜŸáÿßÿ¶Ÿä: {NEW_REPORT_FILE}")
```
